## In this Notebook, I am doing analysis of Covid-19 cases of Pakistan from 27th February till 29th August, 2020. The dataset has been downloaded from the ourworldindata.org website.

```python
# First insert some libraries which I will be using in this Notebook for analysis.
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Now load the dataset with pandas read_csv function, which I named as Pakistan.csv
Pak = pd.read_csv('Pakistan.csv', index_col=0)
```

```python
# Check the first five rows of the dataset.
Pak.head()
```

```python
# Check the complete information of the dataset.
Pak.info()
```

## First let's check some general Information about the distribution of Pakistani Population. The total population of Pakistan is 221,565,832.

```python
fig, ax = plt.subplots()
fig.set_size_inches([9, 7])
median_age = Pak['median_age'].values[0]
aged_65_older = Pak['aged_65_older'].values[0]
aged_70_older = Pak['aged_70_older'].values[0]
diabetes_prevalence = Pak['diabetes_prevalence'].values[0]
female_smokers = Pak['female_smokers'].values[0]
male_smokers = Pak['male_smokers'].values[0]
life_expectancy = Pak['life_expectancy'].values[0]
height = [median_age, aged_65_older, aged_70_older, diabetes_prevalence, female_smokers, male_smokers, life_expectancy]
bars = ('Median Age', 'Aged Older than 65', 'Aged Older than 70', 'Diabetes Prevalence', 'Female Smokers', 'Male Smokers', 'Life Expectancy')
ax.bar(bars, height)
ax.set_xticklabels(bars, rotation=90)
ax.set_title("Some General Information about Pakistan Population")
ax.set_ylabel("Percentage")
plt.show()
```

## As the total number of tests, total number of cases, and total number of deaths from Covid-19 all contains single numbers. Therefore, I am using a bargraph for that.

```python
plt.style.use("bmh")
fig, ax = plt.subplots()
fig.set_size_inches([9, 7])
total_tests = Pak['total_tests'].values[-2]
total_cases = Pak['total_cases'].values[-1]
total_deaths = Pak['total_deaths'].values[-1]
height = [total_tests, total_cases,  total_deaths]
bars = ('Total Tests', 'Total Cases', 'Total Deaths')
ax.bar(bars, height)
ax.set_title("Total Tests vs Total Cases vs Total Deaths in Pakistan")
ax.set_ylabel("Total Count")
ax.set_xticklabels(bars, rotation=90)
plt.show()
```

## Now check the number of tests, cases and deaths per thousand. 

```python
plt.style.use("seaborn-colorblind")
fig, ax = plt.subplots()
fig.set_size_inches([9, 7])
total_tests_per_thousand = Pak['total_tests_per_thousand'].values[-2]
total_cases_per_thousand = (Pak['total_cases_per_million'].values[-1]) / 1000
total_deaths_per_thousand = (Pak['total_deaths_per_million'].values[-1]) / 1000
height = [total_tests_per_thousand, total_cases_per_thousand,  total_deaths_per_thousand]
bars = ('Total Tests per Thousand', 'Total Cases per Thousand', 'Total Deaths per Thousand')
ax.bar(bars, height)
ax.set_xticklabels(bars, rotation=90)
plt.show()
```

## In Pakistan, the first COVID-19 case was registered on 27th February, 2020. Now, lets display the month wise COVID-19 cases in Pakistan. The cases of every month has been displayed in a seperate graph. 

```python
plt.style.use("bmh")
fig, ax = plt.subplots(6,figsize=(15,25), tight_layout=False,constrained_layout=True)

ax[0].plot(Pak['date'], Pak['new_cases'], marker='o', color='blue', label = 'New Cases')
ax[0].plot(Pak['date'], Pak['new_deaths'], marker='x', color='red', label = 'New Deaths')
ax[0].plot(Pak['date'], Pak['new_tests'], marker='v', color='green', label = 'New Tests')
ax[0].set_xlim([0, 33])
ax[0].set_xticklabels(Pak['date'], rotation=90)
ax[0].set_title("COVID-19 Information for the month of February-March")
ax[0].set_xlabel("Dates")
ax[0].set_ylabel("Total Count")
ax[0].legend()

ax[1].plot(Pak['date'], Pak['new_cases'], marker='o', color='blue', label = 'New Cases')
ax[1].plot(Pak['date'], Pak['new_deaths'], marker='x', color='red', label = 'New Deaths')
ax[1].plot(Pak['date'], Pak['new_tests'], marker='v', color='green', label = 'New Tests')
ax[1].set_xlim([34, 63])
ax[1].set_xticklabels(Pak['date'], rotation=90)
ax[1].set_title("COVID-19 Information for the month of April")
ax[1].set_xlabel("Dates")
ax[1].set_ylabel("Total Count")
ax[1].legend()

ax[2].plot(Pak['date'], Pak['new_cases'], marker='o', color='blue', label = 'New Cases')
ax[2].plot(Pak['date'], Pak['new_deaths'], marker='x', color='red', label = 'New Deaths')
ax[2].plot(Pak['date'], Pak['new_tests'], marker='v', color='green', label = 'New Tests')
ax[2].set_xlim([64, 94])
ax[2].set_xticklabels(Pak['date'], rotation=90)
ax[2].set_title("COVID-19 Information for the month of May")
ax[2].set_xlabel("Dates")
ax[2].set_ylabel("Total Count")
ax[2].legend()

ax[3].plot(Pak['date'], Pak['new_cases'], marker='o', color='blue', label = 'New Cases')
ax[3].plot(Pak['date'], Pak['new_deaths'], marker='x', color='red', label = 'New Deaths')
ax[3].plot(Pak['date'], Pak['new_tests'], marker='v', color='green', label = 'New Tests')
ax[3].set_xlim([95, 124])
ax[3].set_xticklabels(Pak['date'], rotation=90)
ax[3].set_title("COVID-19 Information for the month of June")
ax[3].set_xlabel("Dates")
ax[3].set_ylabel("Total Count")
ax[3].legend()

ax[4].plot(Pak['date'], Pak['new_cases'], marker='o', color='blue', label = 'New Cases')
ax[4].plot(Pak['date'], Pak['new_deaths'], marker='x', color='red', label = 'New Deaths')
ax[4].plot(Pak['date'], Pak['new_tests'], marker='v', color='green', label = 'New Tests')
ax[4].set_xlim([125, 155])
ax[4].set_xticklabels(Pak['date'], rotation=90)
ax[4].set_title("COVID-19 Information for the month of July")
ax[4].set_xlabel("Dates")
ax[4].set_ylabel("Total Count")
ax[4].legend()

ax[5].plot(Pak['date'], Pak['new_cases'], marker='o', color='blue', label = 'New Cases')
ax[5].plot(Pak['date'], Pak['new_deaths'], marker='x', color='red', label = 'New Deaths')
ax[5].plot(Pak['date'], Pak['new_tests'], marker='v', color='green', label = 'New Tests')
ax[5].set_xlim([156, 184])
ax[5].set_xticklabels(Pak['date'], rotation=90)
ax[5].set_title("COVID-19 Information for the month of August")
ax[5].set_xlabel("Dates")
ax[5].set_ylabel("Total Count")
ax[5].legend()

plt.show()
```

## Conclusion : As it can be seen from these above six graphs that there was a spike on the number of new cases in May and June but from July and  August the number of new cases and deaths are continuously decreasing. There are many factors for this but the main reason is the rapid increase in recoveries (which was not available in this dataset). However, the reasons for decline in the number of new COVID-19 cases are the imposement of smart lockdown from Pakistani government, other factors such as higher number of youths (which was shown in a bar graph above as the median age) who usually recover very fast and so the corona is not transmitted to higher number of people. One factor is also the increase in temperature (more than 42 degree centigrade in July and August) due to which very less people come outside from their homes, which also plays a significant role in the decline of new COVID-19 cases. Furthermore, the main reasons for less number of deaths in Pakistan from COVID-19 is due to higher number of young populations and less number of smokers and old aged people. There are some missing information (null values) for the number of tests but that are not much important for this analysis.
